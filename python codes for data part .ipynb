```python
# load the data using pandas
file_path = '4.22data.xlsx'
data = pd.read_excel(file_path)
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x='Date',y='Bitcoin Price',data=df,color='blue')
sns.lineplot(x='Date',y='Natural Gas  Price',data=df,color='black')
plt.show()
```

```python
df = data

ax = df.plot(x="Date", y="Gold Price", legend=False,color="red")
ax2 = ax.twinx()
df.plot(x="Date", y="Natural Gas  Price", ax=ax2, legend=False, color="black")
ax.figure.legend()
plt.show()
```
